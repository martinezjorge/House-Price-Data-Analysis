```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
def summarize_data(df):
    categorical = []
    numerical = []
    for column in df.columns:
        print(column)
        if df.dtypes[column] == np.object: # Categorical data
            print(len(np.unique(df[column].tolist())), "unique",column + "\'s")
            print(df[column].value_counts())
            categorical.append(column)
        else:
            print(df[column].describe())
            numerical.append(column)
        print('\n')
    return(numerical,categorical)
```

```python
data = pd.read_csv('train.csv')
data
```

```python
for column in data.columns:
    print(column)
```

```python
data.sample(20)
```

```python
numerical, categorical = summarize_data(data)
numerical = numerical[1:len(numerical)] # shave off the first element in the
numerical.remove('SalePrice')
```

```python
print(numerical)
```

Linear regressions of every numerical piece of data vs. Sale Price

```python
LR = LinearRegression() # instantiate linear regression
for column in numerical:
    test = data[data[column].notnull()]
    x = np.array(test[column])
    y = np.array(test['SalePrice'])
    x = x.reshape(-1,1)
    xtrain, xtest, ytrain, ytest = train_test_split(x,y)
    model=LR.fit(xtrain,ytrain)
    x_pretty=np.linspace(xtest.min(),xtest.max(),xtest.max()-xtest.min())
    y_predicted=LR.predict(x_pretty.reshape(-1,1))
    plt.title(column +' vs. Sale Price')
    plt.xlabel(column)
    plt.ylabel('Sale Price')
    plt.scatter(xtest,ytest)
    plt.plot(x_pretty,y_predicted)
    plt.savefig(column +' vs. Sale Price.png')
            
```
